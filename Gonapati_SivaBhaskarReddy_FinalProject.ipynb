This indicates that a higher review count doesn't necessarily correlate with higher ratings. Interestingly, businesses with a large number of reviews aren't always among the highest-rated, suggesting that customer experiences are varied and not solely dependent on the volume of feedback.

We can also see from that graph that a concentration of businesses with moderate review counts that have average to high star ratings, with few low ratings. This might suggest that while a greater number of reviews doesn't automatically lead to higher ratings, it could be associated with maintaining a certain level of quality, as indicated by the absence of very low ratings among these businesses. Such a pattern could be attributed to several factors such as the nature of the business, the quality of products or services, customer service, and the way businesses handle customer feedback.

The data does not present a straightforward linear correlation between review counts and star ratings, leaving room for the possibility of a more complex relationship. Factors like active engagement with customers, the type of products or services offered, and how businesses respond to both positive and negative reviews could all play a role in shaping this relationship. To understand the dynamics fully, a comprehensive statistical analysis would be needed, possibly including other business attributes and customer interaction metrics.

### **2. Is there a difference in the rating distribution (stars) of food establishments by state? (Hint: use the categories variable to extract the records belonging to food establishments - any place that someone could go to get a freshly prepared meal.)**

```python
#filtering records related to food establishments based on the categories column containing either food or restaurants string
food_establishments = new_yelp_business_df[new_yelp_business_df['categories'].str.contains('Food|Restaurants', case=False)]

#displaying the first 5 rows data
food_establishments.head()
```

```python
#cheking how many states have food businesses 
food_establishment_states = food_establishments['state'].unique()

#displaying the no.of states
print("No.of states with Food Establishments: \n", len(food_establishment_states),'\n')

#printing the state codes to see what are those states
print("State code are: \n", food_establishment_states)
```

--> There are <b>48 state with food establishments.</b> However, from the list of state codes we can observe that, there are some discrepancies in the data where certain state codes appear to be invalid (such as 01, 4, 6, etc.) and some states seem to be outside of the USA. Therefore, I'm <b>considering only the state codes belonging to the USA from the above list.</b>

```python
#list of USA state codes from the above list
us_state_codes = ['AL', 'AZ', 'CA', 'CO', 'IL', 'IN', 'NC', 'NV', 'NY', 'OH', 'OR', 'PA', 'SC', 'VA', 'WI']

#filtering the food_establishments dataframe to include only USA states
food_establishments_us = food_establishments[food_establishments['state'].isin(us_state_codes)]

#displaying sample data
food_establishments_us.head()
```

```python
#checking whether we have only USA states in the dataframe or not
FE_US_states = food_establishments_us['state'].unique()

#displaying the no.of states of USA
print("No.of USA states with Food Establishments: \n", len(FE_US_states),'\n')

#displaying those USA state codes 
print("State code are: \n", FE_US_states)
```

--> Now finding out difference in the rating distribution (stars) of food establishments by state for USA states only.

```python
#grouping data by state and calculate rating distribution
rating_distribution_by_state = food_establishments_us.groupby('state')['stars'].value_counts(normalize=True).rename('percentage').reset_index()

#displaying the top 5 rows of the dataframe
rating_distribution_by_state.head()
```

```python
#plotting the rating distribution by state using seaborn barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='state', y='percentage', hue='stars', data=rating_distribution_by_state)
plt.title('Rating Distribution of Food Establishments by State')
plt.xlabel('State')
plt.ylabel('Percentage')
plt.legend(title='Stars', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()

#displying the graph
plt.show()
```

Yes, there is a significant difference in the distribution of food establishment ratings by state in the United States. Restaurants in the United States have an average rating of 3.5 stars, which indicates a good range of consumer satisfaction.

Oregon has the highest average rating of 4 stars, indicating consistently high levels of satisfaction throughout its various dining outlets. On the other hand, New York has the lowest average rating of 3 stars, suggesting considerably lower satisfaction among its restaurants. States such as California and Virginia boast reputable average scores, signifying widespread customer satisfaction in their food enterprises. However, the factors influencing these scores might vary greatly, including diverse culinary traditions, local tastes, and cuisine varieties.

Overall, the differences in ratings among states showcase the diverse culinary preferences and experiences found across the United States. Based on this results stakeholders can adjust strategies to address to specific local demands, enhancing consumer happiness, and refining overall dining experiences to better align with regional expectations.

### **3. Investigate differences in food establishments with/without a Happy Hour using an appropriate visualization.**

```python
#checking the shape of the dataframe
yelp_business_df.shape
```

```python
#counting the number of null values in the attributes.HappyHour column
null_count = yelp_business_df['attributes.HappyHour'].isnull().sum()

#printing the resluts
print("Number of null values in 'attributes.HappyHour' column:", null_count)
```

```python
#dropping rows with null values in attributes.HappyHour column
happyhour_cleaned_yelp_business_df = yelp_business_df.dropna(subset=['attributes.HappyHour'])

#checking the dataframe shape after dropping null values
print(happyhour_cleaned_yelp_business_df.shape)
```

```python
#counting the number of null values in the attributes.HappyHour column
null_count = happyhour_cleaned_yelp_business_df['attributes.HappyHour'].isnull().sum()

#checking if all the null values are dropped 
print("Number of null values in 'attributes.HappyHour' column:", null_count)
```

```python
#filtering food establishments with Happy Hour which have true value
food_with_happy_hour = happyhour_cleaned_yelp_business_df[(happyhour_cleaned_yelp_business_df['categories'].str.contains('Food|Restaurants')) & 
                                        (happyhour_cleaned_yelp_business_df['attributes.HappyHour'] == True)]

#filtering food establishments without Happy Hour which have false value
food_without_happy_hour = happyhour_cleaned_yelp_business_df[(happyhour_cleaned_yelp_business_df['categories'].str.contains('Food|Restaurants')) & 
                                           (happyhour_cleaned_yelp_business_df['attributes.HappyHour'] == False)]
```

```python
#counting of establishments with and without happy hour
count_with_happy_hour = len(food_with_happy_hour)
count_without_happy_hour = len(food_without_happy_hour)

#calculating average star ratings for food establishments with and without happy hour
avg_stars_with_happy_hour = food_with_happy_hour['stars'].mean()
avg_stars_without_happy_hour = food_without_happy_hour['stars'].mean()

#displaying the average star rating for the food establishments with and without happy hour
print(avg_stars_with_happy_hour)
print(avg_stars_without_happy_hour)

#plotting a bar plot 
plt.figure(figsize=(8, 6))
sns.barplot(x=['With Happy Hour', 'Without Happy Hour'], y=[avg_stars_with_happy_hour, avg_stars_without_happy_hour])
plt.title('Average Star Ratings for Food Establishments with and without Happy Hour')
plt.ylabel('Average Star Rating')


#adding counts value in the bar plot
plt.text(0, avg_stars_with_happy_hour + 0.05, f'Count: {count_with_happy_hour}', ha='center')
plt.text(1, avg_stars_without_happy_hour + 0.05, f'Count: {count_without_happy_hour}', ha='center')

#displaying the bar graph
plt.show()
```

From the graph above, we observe that 5,181 food establishments offer Happy Hour, while 1,220 do not. Interestingly, the average star rating for establishments with Happy Hour is approximately 3.49, slightly lower (though potentially negligible) than that of establishments without Happy Hour, which stands at around 3.64.

From the above numbers we can say that there is no much difference in rating for food establishments with or without happyhour. We may need to consider other factors also may be like customer experiences/satisfaction, quality of food etc., for better understanding.

### 5 (b). Do reviews with exclamation points seem to be either very highly rated or very low rated? Determine the stars distribution as a function of the number of exclamation points used in the review. Draw a suitable plot.

```python
#extracting the number of exclamation points in each review
yelp_review_df['num_exclamation_points'] = yelp_review_df['text'].str.count('!')

#plotting a bar graph for average exclamation points by star rating
exclamation_avg = yelp_review_df.groupby('stars')['num_exclamation_points'].mean()
exclamation_avg.plot(kind='bar', figsize=(12, 8), color='green')
plt.xlabel('Star Rating')
plt.ylabel('Average Number of Exclamation Points')
plt.title('Average Exclamation Points by Star Rating per Review')

#dispolaying the graph
plt.show()
```


From the above graph we can see that the average number of exclamation points for different star ratings are as below:<br>

5-star ratings have an average of 1.75 exclamation points.<br>
1-star ratings have an average of 1.5 exclamation points.<br>
4-star ratings have an average of 1.25 exclamation points.<br>
2 and 3-star ratings have an average of 0.75 exclamation points.<br>


This suggests that higher-rated reviews (5 stars) tend to include a relatively higher number of exclamation points compared to the lower-rated reviews (1 and 4 stars).  Additionally, reviews with 2 or 3 stars use fewer exclamation points on average. In these cases, people don't express extreme positive or negative sentiments with many exclamation points. It's as if the number of exclamation points matches how strongly someone feels about the place they re are reviewing.
